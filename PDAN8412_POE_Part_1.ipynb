# Theory

## 1.	Explanation of Recurrent Neural Network (RNN):
#### A recurrent neural network (RNN) is an artificial neural network used to analyse sequential input. Unlike standard neural networks, RNNs feature connections that create directed cycles, allowing information to be stored. This structure allows RNNs to remember past inputs, making them ideal for applications like language modelling, time series prediction, and sequence labelling, where the order and context of data points are critical. RNNs can deal with variable-length sequences and capture temporal relationships, making them useful for natural language processing and other sequential data tasks.(IBM, n.d.)


## 2.	An explanation of why the dataset is appropriate for processing using a recurrent neural network:
#### The dataset is appropriate for processing with a recurrent neural network (RNN)because RNNs excel at processing sequential data because they maintain the order and context, which is critical for reading and interpreting text. They may detect long-term connections and contextual links within sequences using methods like as LSTM and GRU. The capacity to learn from prior inputs improves the accuracy of jobs like sentiment analysis and spam identification. Furthermore, RNNs can successfully manage temporal information, making them excellent for assessing long-term patterns and trends. Thus, RNNs are ideal for extracting meaningful insights from sequential text data.


##3.	What is (LSTM) model:
#### Long Short-Term Memory (LSTM) networks are artificial recurrent neural networks (RNNs) used in deep learning to handle sequential input. Unlike standard feedforward neural networks, LSTMs have feedback connections that allow them to capture temporal relationships. They are intended to address difficulties like disappearing or ballooning gradients in traditional RNNs, making them ideal for jobs like natural language processing (NLP), speech recognition, and time series forecasting.

####Memory cells in LSTM networks consist of three components: an input gate, a forget gate, and an output gate. The input gate chooses how much of the new data is saved, while the forget gate decides what information to delete. The output gate determines how much of the cell's content is used during the concealed state. These gates allow LSTMs to selectively store and handle information across lengthy sequences, making them more successful at jobs that need long-term dependencies. (Hamad, 2023)

### Mapping the target column to 1s and 0s.

```python
# Convert the target values: map 4 to 1 (positive) and keep 0 as negative
df_pandas['target'] = df_pandas['target'].map({4: 1, 0: 0})

# Check for the conversion
df_pandas['target'].value_counts()  # To confirm the mapping
```

### Creating the text processing model.

```python
# ***************************************************************************************
#    Title: Text Preprocessing in Python
#    Author: Geeksforgeeks
#    Date Published: 02 August 2024
#    Date Retrieved: 30 August 2024
#    Code version: 1
#    Availability: https://www.geeksforgeeks.org/text-preprocessing-in-python-set-1/
#
# ***************************************************************************************

# Define the text processing function
def text_process(text):
    STOPWORDS = set(stopwords.words('english') + ['u', 'ü', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure'])
    porter = PorterStemmer()
    nopunc = ''.join([char for char in text if char not in string.punctuation])
    tokens = word_tokenize(nopunc.lower())
    stemmed_words = [porter.stem(word) for word in tokens if word not in STOPWORDS]
    return ' '.join(stemmed_words)
```

### Applying the text processing on the column 'text' .  

```python
# Apply text processing to the 'text_column' in your DataFrame
df_pandas['cleaned_text'] = df_pandas['text'].apply(text_process)
```

### Importing libraries

```python
# Import necessary libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, f1_score, accuracy_score
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
```

### Setiing up the Tokenization.

```python
# ***************************************************************************************
#    Title: Tokenizer Class Arguments with example
#    Author: Akash Deep
#    Date Published: 22 August 2020
#    Date Retrieved: 30 August 2024
#    Code version: 1
#    Availability: https://medium.com/analytics-vidhya/understanding-nlp-keras-tokenizer-class-arguments-with-example-551c100f0cbd
#
# ***************************************************************************************

# Tokenization
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(df_pandas['text'])

# Convert text to sequences
X = tokenizer.texts_to_sequences(df_pandas['text'])

# Padding sequences to the same length
X = pad_sequences(X, maxlen=100)

# Define the target labels (binary: 0 and 1)
y = df_pandas['target'].values
```

### Creating the LSTM model.

```python
# Define the LSTM model
model = Sequential()

# Add embedding layer
model.add(Embedding(input_dim=5000, output_dim=128, input_length=100))

# Add LSTM layer
model.add(LSTM(units=128, return_sequences=False))

# Add Dropout to prevent overfitting
model.add(Dropout(0.5))

# Add Dense layer for classification
model.add(Dense(units=1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

### Training the LSTM model.

```python
# ***************************************************************************************
#    Title: Machine Learning - Train/Test
#    Author: W3Schools
#    Date Published: 2024
#    Date Retrieved: 30 August 2024
#    Code version: 1
#    Availability: https://www.w3schools.com/python/python_ml_train_test.asp
#
# ***************************************************************************************

# Train the LSTM model
history = model.fit(X, y, epochs=3, batch_size=512, validation_split=0.2)
```

### Evaluating the model.

```python
# Evaluate the model
loss, accuracy = model.evaluate(X, y, verbose=0)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

### Making predictions on the model.

```python
# Predict the labels for the input data
y_pred = (model.predict(X) > 0.5).astype("int32")
# Confusion Matrix
cm = confusion_matrix(y, y_pred)
cm
```
